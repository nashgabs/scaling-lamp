Check that the functions are orthogonal

So $\int_{-l}^{l} \phi_n(x) \phi_m(x) dx = l$
if $n=m$

where $\phi_m = \cos(m \pi x/l)$

Let's do $l=\pi$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create discrete independent variable - we omit the repeating point at 2l by setting endpoint = False
x = np.linspace(-np.pi, np.pi, 10, endpoint=False)
x
```

```python
dx = x[1] - x[0]
dx
```

```python
m = 5
phim = np.cos(m*np.pi*x/np.pi) 
psim = np.sin(m*np.pi*x/np.pi) 
```

```python
plt.plot(x, phim)
plt.plot(x, psim);
```

```python
"""
code to do numerical integration using trapezium rule
Because first and last points (0 and 2l) are the same we can just add the point at 0 and ignore factors of 1/2
"""
my_integral = 0
for datapoint in range(0,len(x)):
    my_integral += psim[datapoint] * phim[datapoint]
my_integral *= dx
my_integral
```
