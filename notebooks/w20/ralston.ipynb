# Problem
Implement Ralston's method in Python, and with it solve the ODE
$$\frac{dy(t)}{dt}=bt-ay(t)=g(t,y)$$
Where $a=8$, $b=1$, $y(t:t=0)=4$, $t_{max}=1$

```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(0, 1+dt, dt)
t.shape
a = 8.0
b = 1.0
g = lambda t, y : (b*t)-(a*y)
```

Evaluate at $\frac{2}{3}$ of the interval
$$
y_{i+1}\approx y_i+\Delta t\cdot(\frac{1}{4}\kappa_1+\frac{3}{4}\kappa_2)$$
Where,
$$\kappa_1=g(t_i,y_i)$$
and
$$\kappa_2=g(t_i+\frac{2}{3}\Delta t,y_i+\frac{2}{3}g(t_i,y_i)\Delta t)$$

```python
k2 = lambda t, y, dt : (b*(t+(2*dt)/3))-(a*(y+g(t,y)*(2*dt)/3))
def ralston(t,dt):
    y = np.zeros(t.shape[0]) # define y-space
    y[0] = 4 # boundary condition
    for n in range(t.shape[0]-1): # for each element in time-space
        y[n+1] = y[n]+dt*((1/4)*g(t[n],y[n])+(3/4)*k2(t[n],y[n],dt)) # ralstons method
    return y
```

Plot the integrated function

```python
a=8
b=1
y0=4
Y =  lambda t : np.exp(-a*t)*(y0+(b/pow(a,2)))+(b*t/a)-(b/pow(a,2))
fig, ax = plt.subplots()
ax.plot(t, ralston(t,dt), color="r", alpha=0.7, label=r'$y(t)=\int \frac{dy(t)}{dt} dt$')
ax.plot(t, Y(t), color="b", alpha=0.7, linestyle='--', label=r'$y(t)=exp(-at)(y(0)+\frac{b}{a^2}+\frac{bt}{a}-\frac{b}{a^2}$')
ax.set_title('Ralston\'s Method')
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.legend(loc="upper left")
```
